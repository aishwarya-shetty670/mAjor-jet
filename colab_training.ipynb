```
from roboflow import Roboflow

rf = Roboflow(api_key="VJH6C9xMrmIJt2pFM8De")
project = rf.workspace("asmaa-rashed-alahmari-pfekk").project("pothole-detection-mj8bb")
version = project.version(1)
dataset = version.download("yolov8")
```

```
from ultralytics import YOLO

# Build a standard detection model
model = YOLO('yolov8n.pt')

# Train the standard detection model
results = model.train(
    data=dataset.location + '/data.yaml',
    epochs=50,
    imgsz=640,
    batch=16,
    name='pothole_yolov8n'
)
```
